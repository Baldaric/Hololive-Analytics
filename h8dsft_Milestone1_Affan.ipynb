# 4. Data Cleaning

Before we start meddling in with data analysis, we first must clean the data. From missing values, from unused columns, etc.

First, I drop tables that I won't use. Which in this case is URL and Video_ID.

```python
# drop the unused columns, URL and Video_ID
df = df.drop('URL', axis=1)
df = df.drop('Video_ID', axis=1)
df.head()
```

Since I want to create 2 new columns based on the creator's name, I check their unique values.

```python
df['Creator'].unique()
```

Here I create a table called 'Debut_year' to provide data how long a streamer's career is.

```python
Debut2017 = ['Sora Tokino','Hololive Production','Hololive English','Hololive ID']
Debut2018 = ['Robokosan','Aki Rosenthal','Fubuki Shirakami','Haato Akai','Matsuri Natsuiro',
            'Miko Sakura','Aqua Minato','Shion Murasaki','Ayame Nakiri','Choco Yuzuki','Subaru Ozora',
            'AZKi','Mio Ookami','Mel Yozora']
Debut2019 = ['Okayu Nekomata','Korone Inugami','Suisei Hoshimachi','Pekora Usada','Rushia Uruha','Marine Houshou',
            'Shiranui Flare','Noel Shirogane','Amane Kanata','Watame Tsunomaki','Towa Tokoyami','Luna Himemori']
Debut2020 = ['Ayunda Risu','Moona Hoshinova','Airani Iofifteen','Lamy Yukihana','Nene Momosuzu','Botan Shishiro','Polka Omaru',
            'Mori Calliope','Takanashi Kiara',"Ninomae Ina'nis",'Gawr Gura','Watson Amelia','Kureiji Ollie','Anya Melfissa','Pavolia Reine']
Debut2021 = ['IRyS','Tsukumo Sana','Ceres Fauna','Ouro Kronii','Nanashi Mumei','Hakos Baelz','Laplus Darkness','Lui Takane','Koyori Hakui',
            'Chloe Sakamata','Iroha Kazama']
Debut2022 = ['Vestia Zeta','Kobo Kanaeru','Kaela Kovalskia',]


ListJoin = []
Streamer = df['Creator']

for i in Streamer:
    if i in Debut2017:
        ListJoin.append(2017)
    elif i in Debut2018:
        ListJoin.append(2018)
    elif i in Debut2019:
        ListJoin.append(2019)
    elif i in Debut2020:
        ListJoin.append(2020)
    elif i in Debut2021:
        ListJoin.append(2021)
    elif i in Debut2022:
        ListJoin.append(2022)
    else:
        ListJoin.append('Unknown')

df['Debut_year'] = ListJoin
```

Then to create column about which creator belongs to which branch.

```python
ListBranch = []
NewBranch = df['Creator']

HololiveEN = ['Mori Calliope','Takanashi Kiara',"Ninomae Ina'nis", 
              'Gawr Gura', 'Watson Amelia', 'IRyS','Ceres Fauna',
              'Ouro Kronii','Nanashi Mumei', 'Hakos Baelz', 'Tsukumo Sana']

HololiveID = ['Ayunda Risu','Moona Hoshinova','Airani Iofifteen','Kureiji Ollie',
              'Anya Melfissa','Pavolie Reine','Vestia Zeta','Kaela Kovalskia','Kobo Kanaeru']

Official = ['Hololive Production', 'Hololive English', 'HololiveID']

for i in NewBranch:
    if i in HololiveEN:
        ListBranch.append('EN')
    elif i in HololiveID:
        ListBranch.append('ID')
    elif i in Official:
        ListBranch.append('Official')
    else:
        ListBranch.append('JP')

df['Hololive_Branch'] = ListBranch
```

Here I read on the internet that separating date and time could lead to headache, so I combined Stream_Start_Date_UST and Stream_Start_Time_UST since they are actually related and can be put on the same column. To take advantage of the datetime format of Pandas.

```python
df['Stream_Start_DateTime_UST'] = pd.to_datetime(df['Stream_Start_Date_UST'] + ' ' + df['Stream_Start_Time_UST'])
```

Then I working on the duration column, I notice that this column should have a format of HH:MM:SS yet some of the rows have format of MM:SS. Therefore, I need to add 00: in front of it before I could change the datatype further.

```python
Add0 = df['duration']

ListAdd0 = []
for i in Add0:
    if len(i.split(':')) == 2:
        ListAdd0.append('00:' + i.replace(' ',''))
    else:
        ListAdd0.append(i)

df['duration'] = ListAdd0
```

Then I drop the unused column that I have combined as Stream_Start_DateTime_UST.

```python
df.drop(columns=['Stream_Start_Date_UST','Stream_Start_Time_UST'],inplace=True)
```

Here I change the data type of duration into timedelta. I use timedelta because if I use datetime it would destroy the data and put it on weird manner even though I already state the format to be %H%M%S. I found the timedelta is more acceptable.

```python
df['duration'] = pd.to_timedelta(df['duration'])
```

Here I want to categorize stream duration between short, normal, and long stream duration.

```python
DurCate = df['duration']

ListCat = []
for i in DurCate:
    seconds = i.total_seconds()
    if seconds <= 3600:
        ListCat.append('Short')
    elif 3600 < seconds <= 7200:
        ListCat.append('Normal')
    elif seconds > 7200:
        ListCat.append('Long')

df['duration_category'] = ListCat
```

Now to clean everything else, first I strip space from 3 columns, SuperChat_total, Avg_viewers, Max_viewers.

```python
df['SuperChat_total'] = df['SuperChat_total'].str.strip()
df['Avg_viewers'] = df['Avg_viewers'].str.strip()
df['Max_viewers'] = df['Max_viewers'].str.strip()
```

Then I delete the ','

```python
df['Avg_viewers'] = df['Avg_viewers'].replace(',','',regex=True)
df['SuperChat_total'] = df['SuperChat_total'].replace(',','',regex=True)
df['Max_viewers'] = df['Max_viewers'].replace(',','',regex=True)
```

Then I noticed that some Max_viewers columns have --- value, I assume the stream is private or membership only. Therefore, we unable to get some data, so I decide to just drop it since it was only a sliver of rows.

```python
df.drop(df.index[df['Max_viewers'] == '---'], inplace = True)
```

Then I change anything else into the desired data type.

```python
df['Avg_viewers'] = df['Avg_viewers'].astype(int)
df['Max_viewers'] = df['Max_viewers'].astype(int)
df['SuperChat_total'] = df['SuperChat_total'].astype(float)
df['Debut_year'] = df['Debut_year'].astype(int)
```

## Inferential: Is there a correlation between creator's long career with the income of super chat?

First we need to state the H0 that need to be rejected and H1 as an alternative.<br>
 - H0 : There are no correlation between creator's long career with the income of super chat.
 - H1 : There are correlation between creator's long career with the income of super chat.

In this hipothesis we use 2 columns, SuperChat_total and Career_length. Career_length data can be obtained by reducing the current year (or in this case the latest data) with their Debut_year

```python
current_year = 2023
df['Career_length'] = current_year - df['Debut_year']
```

Then because we do correlation analysis between 2 numerical data and one of the data is not normal distribution, we will use spearman test to see their correlation.

```python
corr_rho, pval_s = stats.spearmanr(df['Career_length'], df['SuperChat_total'])
print(f"rho-correlation: {corr_rho:.2f}, p-value: {pval_s}")
```

While the p-value approximately -0.0000000000042 in which it indicates that the correlation is statistically insignificant. In other words, the H0 : There are no correlation between creator's long career is rejected. It means that there are correlation between creator's career length and the increase of super chat. Yet since the p-value is so close to zero it is rather very weak correlation.

Proven by the rho correlation which have -0.06, it indicates a very weak negative correlation between the Career_length and SuperChat_total. The values is also close to 0 which suggest that as the career length increase the super chat total doesn't increase in significant way. In conclusion there are correlation but it just not significant enough.

**Conclusion**<br>
The correlation between career length and super chat total is exist albeit almost insignificant. Proven by the p-value of spearman which given the value of approximately -0.0000000000042 that we must reject the H0. But it doesn't mean that the correlation is strong, it is proven by the value of rho-correlation that gave out -0.06 that indicates the correlation is very weak. It means that we can say confidently that new or old talents have the same opportunity in increasing the income of Super Chat.

## Data Visualization

#### Stream timing and schedule consistency affect contribution

I want to know the patterns of super chats in correlation with stream timing. So it is decided that the heatmap table could be the best suited for this.
Using the heatmap table, I want to see at what hour and at what day is the super chat are clustered together.

```python
# first from stream_start_DateTime_UST I extract the day and use Name so it easier to see
df['DayOfWeek'] = df['Stream_Start_DateTime_UST'].dt.day_name()
# then I extract the hour from it
df['Hour'] = df['Stream_Start_DateTime_UST'].dt.hour

# now to put heatmap data inside the variable. Find the superchat total median and grouped by dayofweek and hour.
heatmap_data = df.groupby(['DayOfWeek', 'Hour'])['SuperChat_total'].median().reset_index()

# put variabel for the order of days
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# then from the column of DayOfWeek from heatmap_data, then use it for categorical
heatmap_data['DayOfWeek'] = pd.Categorical(heatmap_data['DayOfWeek'], categories=ordered_days, ordered=True)

# now to creat the density heatmap
fig = px.density_heatmap(
    # data that we have been made
    heatmap_data, 
    # column used for X axis
    x="DayOfWeek", 
    # column used for Y axis
    y="Hour", 
    # column that will be stacked between X and Y axis
    z="SuperChat_total",
    title="SuperChat Contributions by Hour and Day of the Week",
    color_continuous_scale="YlGnBu"
)

fig.show()
```

**Conclusion**<br>
Based on the table, we can conlcude that consistency is one of the key to increase super chat. When we track the behavior of super chat in hours per day in one week, everyday have blue colored cell that indicates it have high sum of super chat. Which mean everyday is a potential of income of super chat.

#### High Average Viewers lead to more contribution

Logically speaking, high average views could lead more super chat income since this is the rough number of how many viewers who watched the stream from start to finish. Viewers who watched the whole stream could potentially give more super chat to engage more with their streamer.

```python
# figure, axis
fig, ax = plt.subplots()

ax.set_title('High Average Viewers have More Contribution')

sns.regplot(data=df, x='Avg_viewers', y='SuperChat_total', scatter=False, ax=ax, color='red')
sns.scatterplot(data=df, x='Avg_viewers', y='SuperChat_total', ax=ax)
plt.show()
```

**Conclusion**<br>
Based on this chart, we can concluded that high average viewers doesn't necessarily increase the super chat income. It safe to say that viewers who constantly give super chat are the one who appear early while the creator slowly gather their viewers and reach peak viewers. Rather than focused on increasing average viewers it is better to focused on viewers who get in the stream early.

#### Stream Duration affects the Super Chat income

Logically speaking, long duration of stream can give more opportunities for viewers to give super chat. While short stream may have small window of opportunity to do that. Correlation between these argument needs to be proven by data.

```python
# first I convert the duration into only hour
df['Stream_Duration_Hours'] = df['duration'] / np.timedelta64(1, 'h')
# then I count the income per hour by divide total super chat with hour duration
df['Income_per_Hour'] = df['SuperChat_total'] / df['Stream_Duration_Hours']
fig, ax = plt.subplots()

ax.set_title('Income Efficiency vs Duration')
sns.regplot(data=df, x='Stream_Duration_Hours', y='Income_per_Hour', scatter=False, ax=ax, color='red')
sns.scatterplot(data=df, x='Stream_Duration_Hours', y='Income_per_Hour', ax=ax)
plt.show()
```

**Conclusion**<br>
Based on this data, we can conclude that longer stream is less efficient and do not proportionally increase income. A normal 1 to 2 hour is the best duration considering the average earning. Even though there are several super chats that appear late in the stream, the bulk of super chat is exist at the early time of streaming. We may need to push creators to have more consistent schedule with low duration stream for at least 1 hour.

#### Top Earners Creators Have More Stream Hour than Others

Does top earners do more stream than others? We need to compare their total stream along with their total earnings.

```python
CreatorEarning = df['SuperChat_total'].groupby(df['Creator']).median().reset_index()
CreatorEarning = CreatorEarning.rename(columns={"SuperChat_total": "SuperChat_Median"})

CreatorStream = df['duration'].groupby(df['Creator']).sum().reset_index()
CreatorStream = CreatorStream.rename(columns={'duration':'Total_Stream'})
```

```python
# Merge the two DataFrames on the 'Creator' column
combined_df = pd.merge(CreatorEarning, CreatorStream, on="Creator")

# Melt the DataFrame to have 'Metric' as a column
melted_df = combined_df.melt(
    id_vars="Creator", 
    value_vars=["SuperChat_Median", "Total_Stream"],
    var_name="Metric", 
    value_name="Value"
)

# Create the grouped bar chart
fig = px.bar(
    melted_df,
    x="Creator",
    y="Value",
    color="Metric",
    barmode="group",
    title="Comparison of Median SuperChat and Total Stream Hours by Creator",
    labels={"Value": "Value", "Metric": "Metric", "Creator": "Creator"},
    template="plotly"
)

fig.show()
```

**Conclusion**<br>
Based on this box plot, we can colcluded that the top earners have pretty much the same outliers, means that basically they have the same stream hours. But seeing the mean of each box plot we can concluded that the top earners have more consistent stream hours. Basically they streaming more than others albeit they have no significant difference.
